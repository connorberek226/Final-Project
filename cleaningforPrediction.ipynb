```python
import pandas as pd
from datetime import datetime, timedelta
import numpy as np
```

```python
#Disaster & HPI Cleaning
hpi_disaster_df = pd.read_csv("Xi/states_HPI_disa_weather_final2.csv")

hpi_df = hpi_disaster_df[["State", "Year", "Total_Disaster", "HPI"]].dropna()

hpi_disaster_df

ca_df = hpi_df.loc[hpi_df["State"] == "CA"].copy()
fl_df = hpi_df.loc[hpi_df["State"] == "FL"].copy()
il_df = hpi_df.loc[hpi_df["State"] == "IL"].copy()

ca_df["State"] = "California"
fl_df["State"] = "Florida"
il_df["State"] = "Illinois"

ca_hpi = ca_df[["State", "Year", "HPI"]]
fl_hpi = fl_df[["State", "Year", "HPI"]]
il_hpi = il_df[["State", "Year", "HPI"]]

il_hpi = il_hpi.loc[il_hpi["Year"] >= 1996]
```

```python
frames = [ca_hpi, fl_hpi, il_hpi]

states_hpi = pd.concat(frames)

states_hpi

states_hpi.to_csv("Resources/StateHPIByYear.csv")
```
