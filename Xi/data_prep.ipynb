```python
florida_df['State'] = 'FL'
florida_df.head()
```

```python
california_df['State'] = 'CA'
california_df.head()
```

```python
fl_Cali =florida_df.append(california_df)
```

```python
HPI_csv = "data/HPI_by_county.csv"
HPI_df = pd.read_csv(HPI_csv)
HPI_df
```

```python
ca = HPI_df.loc[(HPI_df["State"] == "CA")]
ca
```

```python
fl = HPI_df.loc[(HPI_df["State"] == "FL")]
fl
```

```python
fl_Cali_HPI =fl.append(ca)
fl_Cali_HPI
```

```python
HPI_Final = fl_Cali_HPI[['State', 'County', 'Year', 'HPI']]
HPI_Final
```

```python
HPI_Final['HPI'] = pd.to_numeric(HPI_Final['HPI'], errors='coerce')
```

```python
HPI_Final.dtypes
```

```python
avg_HPI=HPI_Final.groupby(['State','Year'])['HPI'].mean()         
avg_HPI
```

```python
avg_HPI_final = avg_HPI.to_frame()
avg_HPI_final
```

```python
states_avg_HPI  = pd.merge(left=fl_Cali, right=avg_HPI_final, on=['State','Year'], how='inner')
states_avg_HPI
```

```python
states_avg_HPI.to_csv('states_avgHPI.csv', index=False)
```
