```python
 %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
weather = pd.read_csv("Resources/StateWeatherByYearPredicted2.csv")
disaster = pd.read_csv("Resources/StateDisasterByYearPredicted.csv")
hpi = pd.read_csv("Resources/StateHPIByYearPredicted.csv")

hpi
```

```python
weather_disaster = weather.merge(disaster, how="outer", on=["State", "Year"])

weather_disaster.head(25)
```

```python
weather_disaster_hpi = weather_disaster.merge(hpi, how="outer", on=["State", "Year"])

weather_disaster_hpi = weather_disaster_hpi[["Year", "State", "Average Weather (F, Predicted)", "Total_Disasters (Predicted)", "HPI (Predicted)"]]

weather_disaster_hpi
```

```python
# Assign the data to X and y

X = weather_disaster_hpi[["Average Weather (F, Predicted)", "Total_Disasters (Predicted)"]]
y = weather_disaster_hpi["HPI (Predicted)"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
# Use train_test_split to create training and testing data

### BEGIN SOLUTION
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

### END SOLUTION
```

```python
 # Create the model using LinearRegression

### BEGIN SOLUTION
from sklearn.linear_model import LinearRegression
model = LinearRegression()
### END SOLUTION
```

```python
# Fit the model to the training data and calculate the scores for the training and testing data

### BEGIN SOLUTION
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

### END SOLUTION 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```
